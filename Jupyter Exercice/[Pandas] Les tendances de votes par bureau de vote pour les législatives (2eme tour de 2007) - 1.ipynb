```python
import pandas as pd
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('EXCEL/geo_bureaux_de_vote_md.xlsx')
```

```python
# To make thie excel file, we have to make a candidates list by 'commune code' and 'code bureau de vote' manually using excel function
excel_votes_bureaux_de_vote = pd.ExcelFile('EXCEL/Les tendances de votes par bureau de vote pour les législatives (2eme tour de 2007).xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_votes = pd.read_excel(excel_votes_bureaux_de_vote)
```

```python
df_geo.rename(columns={'code':'N° de bureau de vote'}, inplace=True)
df_geo
```

```python
df_votes.rename(columns={'Code INSEE':'commune_code'}, inplace=True)
df_votes
```

```python
# Inner Join
df_INNER_JOIN = pd.merge(df_geo, df_votes, on=['commune_code','N° de bureau de vote'])
df_INNER_JOIN.to_excel('EXCEL/Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2007).xlsx')
```

```python
df_INNER_JOIN = pd.read_excel('EXCEL/Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2007).xlsx')
```

```python
df=df_INNER_JOIN.groupby(['commune_code', 'N° de bureau de vote'])['Nombre de voix du candidat'].idxmax()
# df_votes.groupby(['Code Commune (Insee Code)', 'N° de bureau de vote'])['Nombre de voix du candidat'].max()

list_index_le_plus_vote=df.to_frame()['Nombre de voix du candidat'].tolist()


df_le_plus_vote= pd.DataFrame(df_INNER_JOIN,
                  index=list_index_le_plus_vote)
```
