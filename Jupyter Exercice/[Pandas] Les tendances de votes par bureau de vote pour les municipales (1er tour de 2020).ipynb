```python
import pandas as pd
import datetime 
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('geo_bureaux_de_vote_md.xlsx')
```

```python
excel_votes_bureaux_de_vote = pd.ExcelFile('Les tendances de votes par bureau de vote pour les municipales (1er tour de 2020)_md.xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_votes = pd.read_excel(excel_votes_bureaux_de_vote)
```

```python
df_geo.rename(columns={'code':'Code B.Vote'}, inplace=True)
df_geo
```

```python
df_votes.rename(columns={'Code du département':'commune_code'}, inplace=True)
df_votes
```

```python
# Inner Join
df_INNER_JOIN = pd.merge(df_geo, df_votes, on=['commune_code','Code B.Vote'])
```

```python
df_INNER_JOIN.to_excel('Les tendances de votes avec cordonnées pour les municipales (1er tour de 2020).xlsx')
```
