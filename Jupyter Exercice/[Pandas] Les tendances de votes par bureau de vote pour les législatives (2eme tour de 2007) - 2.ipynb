```python
import pandas as pd
```

```python
# Open EXCEL/Les tendances de votes par bureau de vote pour les législatives (2eme tour de 2007).xlsx and Then With Excel Function
# Make a candidates list horizontally manually directly in excel file with function IF & AND
# Then Save as like that
excel_bureaux_de_vote=pd.ExcelFile('DataFrame pour les législatives (2eme tour de 2007)/[Pandas] 1,2/1) Les tendances de votes par bureau de vote pour les législatives (2eme tour de 2007)_candidates_list.xlsx')
```

```python
df_commune = pd.read_excel(excel_bureaux_de_vote)
```

```python
df_commune
```

```python
# Here nth(-1) : last
df_groupby_last=df_commune.groupby(['commune_code', 'N° de bureau de vote','longitude','latitude','Inscrits','Votants','Exprimés'])[['Nom du candidat','Prénom du candidat','Code nuance du candidat','Nombre de voix du candidat','% Voix/Exp','list_candidates']].nth(-1)
```

```python
df_groupby_last
```

```python
# For get the information of candidates horizontally, we save df_groupby_last. And then, Manually we abstract and make the candidates list (CSV)
df_groupby_last.to_excel('DataFrame pour les législatives (2eme tour de 2007)/[Pandas] 1,2/2)Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2007)_groupby_last.xlsx')
```

```python
excel_votes_bureaux_de_vote_final = pd.ExcelFile('DataFrame pour les législatives (2eme tour de 2007)/[Pandas] 1,2/5) Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2007)_final_before_MAX.xlsx')
```

```python
# Inner Join 
# 'tendances de votes avec cordonnées pour les législatives (2eme tour de 2007)_final.xlsx'
# 'df_le_plus_vote.xlsx'
# Add le plus voté as column in tendances de votes avec cordonnées pour les législatives (2eme tour de 2007)_final
```

```python
excel_le_plus_vote=pd.ExcelFile('DataFrame pour les législatives (2eme tour de 2007)/[Pandas] 1,2/df_le_plus_vote.xlsx')
```

```python
df_votes_final=pd.read_excel(excel_votes_bureaux_de_vote_final)
df_le_plus_vote=pd.read_excel(excel_le_plus_vote)
```

```python
df_INNER_JOIN = pd.merge(df_votes_final,df_le_plus_vote, on=['commune_code','N° de bureau de vote'])

df_INNER_JOIN
```

```python
df_INNER_JOIN.to_excel('DataFrame pour les législatives (2eme tour de 2007)/Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2007)_final.xlsx')
```
