```python
# Le Plus Voté (Nom, Prénom)
import random
import itertools

list_nom=[]
list_prenom=[]

list_nom.append(df_commune['NOM_candidat_le_plus_voté'].tolist())
list_prenom.append(df_commune['PRENOM_candidat_le_plus_voté'].tolist())

# Contained 2nd Array Dimension in List -> Remove 2nd Array, and take the elements in list 
list_nom = sum(list_nom, [])
list_prenom = sum(list_prenom, [])

# Remove Duplicated elements
dict_nom = dict.fromkeys(list_nom)
list_candidat = list(set(zip(list_nom, list_prenom)))
    
# Color for Each Politic Party
# for key, value in dict_nom.items():
#     dict_nom[key]="#%06x" % random.randint(0, 0xFFFFFF)

dict_nom['MACRON']="#FFB400"
dict_nom['LE PEN']="#2A25C7"
```

```python
# Version 1 (Plot)
import folium
from folium import plugins

map = folium.Map(location=[46.91160617052484, 2.153649265809747]
                       , zoom_start=5.5)


def colorByName(Nom):
    
    return dict_nom[Nom]


df_commune.apply(lambda row: folium.Circle(location=[row.loc['latitude'], row.loc['longitude']],
                                             radius=100,
                                             popup=row.loc['NOM_candidat_le_plus_voté'],
                                             fill=True, 
                                             tooltip="""
                                                     Nom : <b>"""+ str(row['NOM_candidat_le_plus_voté']) +"""</b><br>
                                                     Prénom : <b>"""+ str(row['PRENOM_candidat_le_plus_voté']) +"""</b><br>
                                                     % Voix/Exp : <b>"""+ str(row['% Voix/Exp_le_plus_voté']) +"""</b><br>
                                                     
                                                    
                                                     
                                                     """, 
                                             color=colorByName(row.loc['NOM_candidat_le_plus_voté']), 
                                             fill_color=colorByName(row['NOM_candidat_le_plus_voté']),
                                             opacity=0.8,
                                             fill_opacity=0.7)
                                             .add_to(map), axis=1)    

# map
```
