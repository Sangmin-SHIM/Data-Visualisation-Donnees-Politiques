```python
import pandas as pd
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('geo_bureaux_de_vote_md.xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_geo
```

```python
excel_votes_bureaux_de_vote = pd.ExcelFile('Les tendances de votes par bureau de vote pour les présidentielles (2ème tour de 2017)_md.xlsx')
```

```python
df_votes = pd.read_excel(excel_votes_bureaux_de_vote)
```

```python
df_geo.rename(columns={'code':'Code du b.vote'}, inplace=True)
df_votes.rename(columns={'Code du département':'commune_code'}, inplace=True)
```

```python
# Inner Join
df_INNER_JOIN = pd.merge(df_geo, df_votes, on=['commune_code','Code du b.vote'])
```

```python
df_INNER_JOIN.to_excel('Les tendances de votes avec cordonnées (2ème tour de 2017).xlsx')
```
