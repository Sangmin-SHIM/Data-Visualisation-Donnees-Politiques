```python
import pandas as pd
```

```python
excel_final=pd.ExcelFile('DataFrame pour les législatives (1er tour de 2012)/3)Les tendances de votes avec cordonnées pour les législatives (1er tour de 2012)_final.xlsx')
```

```python
df_commune = pd.read_excel(excel_final)
```

```python
# number = 26, 25, 24, 23, ... , 7

def makeDataFrame(number):
    global df_commune
    
    # 1) Make a new DataFrame 
    index_list = df_commune[df_commune['Nom du candidat.'+str(number)].notnull()].index.tolist()
    globals()['df_voix_'+str(number)] = pd.DataFrame(df_commune, index=index_list)
    
    # 2) df_commune - df_new_dataframe
    df = pd.concat([df_commune, globals()['df_voix_'+str(number)]])
    df = df.drop_duplicates(keep=False)
    df_commune = df
    
    # 3) Make Excel File
    globals()['df_voix_'+str(number)].to_excel('DataFrame pour les législatives (1er tour de 2012)/df_voix_'+str(number)+'.xlsx')
```

```python
for number in range(26,5,-1):
    if number == 6:
        df_commune.to_excel('DataFrame pour les législatives (1er tour de 2012)/df_voix_6_to_0.xlsx')
    makeDataFrame(number)
```
