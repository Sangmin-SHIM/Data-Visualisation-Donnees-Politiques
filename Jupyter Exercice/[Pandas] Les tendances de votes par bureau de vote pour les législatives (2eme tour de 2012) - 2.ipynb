```python
excel_votes_bureaux_de_vote_final = pd.ExcelFile('DataFrame pour les législatives (2eme tour de 2012)/[Pandas] 1,2/5) Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2012)_final.xlsx')
```

```python
excel_le_plus_vote=pd.ExcelFile('DataFrame pour les législatives (2eme tour de 2012)/[Pandas] 1,2/df_le_plus_vote.xlsx')
```

```python
df_votes_final=pd.read_excel(excel_votes_bureaux_de_vote_final)
df_le_plus_vote=pd.read_excel(excel_le_plus_vote)
```

```python
df_INNER_JOIN = pd.merge(df_votes_final,df_le_plus_vote, on=['commune_code','N° de bureau de vote'])

df_INNER_JOIN
```

```python
df_INNER_JOIN.to_excel('DataFrame pour les législatives (2eme tour de 2012)/Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2012)_final.xlsx')
```
