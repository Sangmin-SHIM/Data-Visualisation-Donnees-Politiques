```python
import pandas as pd
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('EXCEL/geo_bureaux_de_vote_md.xlsx')
```

```python
excel_votes_commune = pd.ExcelFile('EXCEL/Les tendances de votes par commune pour les présidentielle (1er tour de 2022).xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_votes = pd.read_excel(excel_votes_commune)
```

```python
df_geo
```

```python
df_geo=df_geo[df_geo["code"]==1]
df_geo
```

```python
df_votes
```

```python
# Inner Join
df_INNER_JOIN = pd.merge(df_geo, df_votes, on=['commune_code'])
df_INNER_JOIN
```

```python
df_INNER_JOIN.to_excel('EXCEL/Les tendances de votes avec cordonnées pour les présidentielles (1er tour de 2022).xlsx')
```
