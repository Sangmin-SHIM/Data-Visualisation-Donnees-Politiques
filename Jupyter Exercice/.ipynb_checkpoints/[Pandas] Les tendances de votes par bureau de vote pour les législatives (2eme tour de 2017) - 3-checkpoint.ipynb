```python
import pandas as pd

# Ignore SettingWithCopyWarning Pandas
pd.set_option('mode.chained_assignment', None)
```

```python
for number in range(7,25):
    if number == 7:
        excel_7_0=pd.ExcelFile('DataFrame pour les législatives (1er tour de 2017)/df_voix_7_to_0.xlsx')
        df_voix_7_0 = pd.read_excel(excel_7_0)
    else:
        excel_name = pd.ExcelFile('DataFrame pour les législatives (1er tour de 2017)/df_voix_'+str(number)+'.xlsx')
        globals()['df_voix_'+str(number)] = pd.read_excel(excel_name)
```

```python
for number in range(7,25):
    if number==7:
        # 1) Initialisation = Make new columns
        df_voix_7_0["% Voix/Exp_le_plus_voté"]=""
        df_voix_7_0["NOM_candidat_le_plus_voté"]=""
        df_voix_7_0["PRENOM_candidat_le_plus_voté"]=""
        df_voix_7_0["NUANCE_candidat_le_plus_voté"]=""

        # 2) Make list for new column names
        for i in range(0,25):
            if i == 0:
                voix_exp_list = ["% Voix/Exp"]
                voix_nom_list = ["Nom"]
                voix_prenom_list = ["Prénom"]
                voix_nuance_list = ["Nuance"]
            else:        
                voix_exp_list.append("% Voix/Exp."+str(i))
                voix_nom_list.append("Nom."+str(i))
                voix_prenom_list.append("Prénom."+str(i))
                voix_nuance_list.append("Nuance."+str(i))


        # 3) Make each DataFrame with values "exp", "nom", "prénom", "nuance"    
        df_voix_exp=df_voix_7_0[voix_exp_list]
        df_voix_nom=df_voix_7_0[voix_nom_list]
        df_voix_prenom=df_voix_7_0[voix_prenom_list]
        df_voix_nuance=df_voix_7_0[voix_nuance_list]

        # 4) In each new list, the most voted Candidat(Nom, Prénom, Nuance) information all listed
        list_nom_le_plus_vote=[]
        list_prenom_le_plus_vote=[]
        list_nuance_le_plus_vote=[]

        for i in range(0,len(df_voix_7_0)):
             list_nom_le_plus_vote.append(df_voix_7_0[df_voix_exp.idxmax(axis=1)[i].replace('% Voix/Exp','Nom')][i])
             list_prenom_le_plus_vote.append(df_voix_7_0[df_voix_exp.idxmax(axis=1)[i].replace('% Voix/Exp','Prénom')][i])
             list_nuance_le_plus_vote.append(df_voix_7_0[df_voix_exp.idxmax(axis=1)[i].replace('% Voix/Exp','Nuance')][i])

        # 5) Insert data in list to Original DataFrame
        df_voix_7_0["% Voix/Exp_le_plus_voté"]=df_voix_exp.max(axis=1).tolist()
        df_voix_7_0['NOM_candidat_le_plus_voté']=list_nom_le_plus_vote
        df_voix_7_0['PRENOM_candidat_le_plus_voté']=list_prenom_le_plus_vote
        df_voix_7_0['NUANCE_candidat_le_plus_voté']=list_nuance_le_plus_vote
        
        # 6) Save In Excel
        df_voix_7_0.to_excel('DataFrame pour les législatives (1er tour de 2017)/df_voix_7_to_0.xlsx')
    else:
        # 1) Initialisation = Make new columns
        globals()['df_voix_'+str(number)]["% Voix/Exp_le_plus_voté"]=""
        globals()['df_voix_'+str(number)]["NOM_candidat_le_plus_voté"]=""
        globals()['df_voix_'+str(number)]["PRENOM_candidat_le_plus_voté"]=""
        globals()['df_voix_'+str(number)]["NUANCE_candidat_le_plus_voté"]=""

        # 2) Make list for new column names
        for i in range(0,25):
            if i == 0:
                voix_exp_list = ["% Voix/Exp"]
                voix_nom_list = ["Nom"]
                voix_prenom_list = ["Prénom"]
                voix_nuance_list = ["Nuance"]
            else:        
                voix_exp_list.append("% Voix/Exp."+str(i))
                voix_nom_list.append("Nom."+str(i))
                voix_prenom_list.append("Prénom."+str(i))
                voix_nuance_list.append("Nuance."+str(i))


        # 3) Make each DataFrame with values "exp", "nom", "prénom", "nuance"    
        df_voix_exp=globals()['df_voix_'+str(number)][voix_exp_list]
        df_voix_nom=globals()['df_voix_'+str(number)][voix_nom_list]
        df_voix_prenom=globals()['df_voix_'+str(number)][voix_prenom_list]
        df_voix_nuance=globals()['df_voix_'+str(number)][voix_nuance_list]


        # 4) In each new list, the most voted Candidat(Nom, Prénom, Nuance) information all listed
        list_nom_le_plus_vote=[]
        list_prenom_le_plus_vote=[]
        list_nuance_le_plus_vote=[]

        for i in range(0,len(globals()['df_voix_'+str(number)])):
             list_nom_le_plus_vote.append(globals()['df_voix_'+str(number)][df_voix_exp.idxmax(axis=1)[i].replace('% Voix/Exp','Nom')][i])
             list_prenom_le_plus_vote.append(globals()['df_voix_'+str(number)][df_voix_exp.idxmax(axis=1)[i].replace('% Voix/Exp','Prénom')][i])
             list_nuance_le_plus_vote.append(globals()['df_voix_'+str(number)][df_voix_exp.idxmax(axis=1)[i].replace('% Voix/Exp','Nuance')][i])

        # 5) Insert data in list to Original DataFrame
        globals()['df_voix_'+str(number)]["% Voix/Exp_le_plus_voté"]=df_voix_exp.max(axis=1).tolist()
        globals()['df_voix_'+str(number)]['NOM_candidat_le_plus_voté']=list_nom_le_plus_vote
        globals()['df_voix_'+str(number)]['PRENOM_candidat_le_plus_voté']=list_prenom_le_plus_vote
        globals()['df_voix_'+str(number)]['NUANCE_candidat_le_plus_voté']=list_nuance_le_plus_vote
        
        # 6) Save In Excel
        globals()['df_voix_'+str(number)].to_excel('DataFrame pour les législatives (1er tour de 2017)/df_voix_'+str(number)+'.xlsx')
```
