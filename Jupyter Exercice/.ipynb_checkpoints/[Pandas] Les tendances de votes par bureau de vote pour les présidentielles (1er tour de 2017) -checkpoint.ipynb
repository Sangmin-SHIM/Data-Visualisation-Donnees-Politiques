```python
import pandas as pd
import datetime 
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('geo_bureaux_de_vote_md.xlsx')
```

```python
excel_votes_bureaux_de_vote = pd.ExcelFile('Les tendances de votes par bureau de vote pour les présidentielles (1er tour de 2017)_md.xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_votes = pd.read_excel(excel_votes_bureaux_de_vote)
```

```python
df_geo.rename(columns={'code':'Code du b.vote'}, inplace=True)
df_geo
```

```python
df_votes.rename(columns={'Code du département':'commune_code'}, inplace=True)
df_votes
```

```python
# Inner Join
df_INNER_JOIN = pd.merge(df_geo, df_votes, on=['commune_code','Code du b.vote'])
```

```python
df_INNER_JOIN.to_excel('Les tendances de votes avec cordonnées (1er tour de 2017).xlsx')
```

```python
excel_final=pd.ExcelFile('Les tendances de votes avec cordonnées (1er tour de 2017).xlsx')
```

```python
df_commune = pd.read_excel(excel_final)
```

```python
# Data Check for [longitude, latitude]
# ------------------------------------

df_commune[df_commune.latitude.apply(type) == datetime.datetime]
df_commune[df_commune.longitude.apply(type) == datetime.datetime]

df_commune[df_commune.latitude.apply(type) == str]
df_commune[df_commune.longitude.apply(type) == str]

# Le Plus Voté - Nom, Prénom et % Voix / Ins
# ------------------------------------------

df_voix_exp = df_commune[['% Voix/Exp', '% Voix/Exp.1', '% Voix/Exp.2' , '% Voix/Exp.3' , '% Voix/Exp.4' , '% Voix/Exp.5' , '% Voix/Exp.6' , '% Voix/Exp.7', '% Voix/Exp.8', '% Voix/Exp.9', '% Voix/Exp.10']]

class Candidat:
    def __init__(self, nom, prenom):
        self.nom = nom
        self.prenom = prenom

candidat=[]

candidat.append(Candidat('DUPONT-AIGNAN','Nicolas'))
for i in range(2,12):
    candidat.append(Candidat(df_commune["Nom."+str(i-1)][0], df_commune["Prénom."+str(i-1)][0]))

df_commune['NOM_candidat_le_plus_voté']=df_voix_exp.idxmax(axis=1).str.replace('% Voix/Exp','Nom')
df_commune['PRENOM_candidat_le_plus_voté']=df_voix_exp.idxmax(axis=1).str.replace('% Voix/Exp','Prénom')

df_commune['% Voix/Exp_le_plus_voté']=df_voix_exp.max(axis=1)

df_commune.loc[(df_commune.NOM_candidat_le_plus_voté =='Nom'),'NOM_candidat_le_plus_voté'] = candidat[0].nom
df_commune.loc[(df_commune.PRENOM_candidat_le_plus_voté =='Prénom'),'PRENOM_candidat_le_plus_voté'] = candidat[0].prenom
for i in range(2,12):
    df_commune.loc[(df_commune.NOM_candidat_le_plus_voté =='Nom.'+str(i-1),'NOM_candidat_le_plus_voté')] = candidat[i-1].nom
    df_commune.loc[(df_commune.PRENOM_candidat_le_plus_voté =='Prénom.'+str(i-1),'PRENOM_candidat_le_plus_voté')] = candidat[i-1].prenom


# df.loc[(df.Event == 'Dance'),'Event']='Hip-Hop'
# df_commune.to_excel('Les tendances de votes avec cordonnées (1er tour de 2017).xlsx')
```
