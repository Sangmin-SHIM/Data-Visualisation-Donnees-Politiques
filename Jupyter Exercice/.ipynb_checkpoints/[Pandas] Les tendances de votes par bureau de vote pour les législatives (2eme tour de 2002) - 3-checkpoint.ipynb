```python
import pandas as pd
```

```python
# Before execute this code, edit manually this excel like 'Changing Column name - NOM_le_plus_voté' and '_x, _y' delete ...
excel_final=pd.ExcelFile('DataFrame pour les législatives (2eme tour de 2002)/Les tendances de votes avec cordonnées pour les législatives (2eme tour de 2002)_final.xlsx')
```

```python
df_commune = pd.read_excel(excel_final)
```

```python
# number = 3, (2 ~ 0)

def makeDataFrame(number):
    global df_commune
    
    # 1) Make a new DataFrame 
    index_list = df_commune[df_commune['Nom du candidat.'+str(number)].notnull()].index.tolist()
    globals()['df_voix_'+str(number)] = pd.DataFrame(df_commune, index=index_list)
    
    # 2) df_commune - df_new_dataframe
    df = pd.concat([df_commune, globals()['df_voix_'+str(number)]])
    df = df.drop_duplicates(keep=False)
    df_commune = df
    
    # 3) Make Excel File
    globals()['df_voix_'+str(number)].to_excel('DataFrame pour les législatives (2eme tour de 2002)/df_voix_'+str(number)+'.xlsx')
```

```python
for number in range(3,1,-1):
    if number == 2:
        df_commune.to_excel('DataFrame pour les législatives (2eme tour de 2002)/df_voix_2_to_0.xlsx')
        break
    makeDataFrame(number)
```
