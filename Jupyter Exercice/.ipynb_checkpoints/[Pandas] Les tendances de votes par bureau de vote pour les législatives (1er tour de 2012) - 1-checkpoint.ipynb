```python
import pandas as pd
import numpy as np
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('geo_bureaux_de_vote_md.xlsx')
```

```python
excel_votes_bureaux_de_vote = pd.ExcelFile('Les tendances de votes par bureau de vote pour les législatives (1er tour de 2012).xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_votes = pd.read_excel(excel_votes_bureaux_de_vote)
```

```python
df_geo.rename(columns={'code':'Code du b.vote'}, inplace=True)
df_geo
```

```python
def func(x):
    df={}
    df['Nom du candidat.1']=x['Nom du candidat']
    
    return pd.Series(df, index=['Nom du candidat.1'])

df_votes.groupby(['Code Commune (Insee Code)', 'N° de bureau de vote']).apply(func).reset_index()
```
