```python
import pandas as pd
```

```python
excel_geo_bureaux_de_vote = pd.ExcelFile('geo_bureaux_de_vote_md.xlsx')
```

```python
excel_votes_bureaux_de_vote = pd.ExcelFile('Les tendances de votes par bureau de vote pour les législatives (1er tour de 2012).xlsx')
```

```python
excel_le_plus_vote=pd.ExcelFile('DataFrame pour les législatives (1er tour de 2012)/df_le_plus_vote.xlsx')
```

```python
excel_votes_bureaux_de_vote_final = pd.ExcelFile('DataFrame pour les législatives (1er tour de 2012)/3)Les tendances de votes avec cordonnées pour les législatives (1er tour de 2012)_final.xlsx')
```

```python
df_geo = pd.read_excel(excel_geo_bureaux_de_vote)
```

```python
df_excel_le_plus_vote = pd.read_excel(excel_le_plus_vote)
```

```python
df_votes = pd.read_excel(excel_votes_bureaux_de_vote)
```

```python
df_votes_final = pd.read_excel(excel_votes_bureaux_de_vote_final)
```

```python
df_geo.rename(columns={'code':'N° de bureau de vote'}, inplace=True)
df_geo
```

```python
df_votes.rename(columns={'Code Commune (Insee Code)':'commune_code'}, inplace=True)
df_votes
```

```python
# Inner Join
df_INNER_JOIN = pd.merge(df_geo, df_votes, on=['commune_code','N° de bureau de vote'])
```

```python
# df_INNER_JOIN
```

```python
df_INNER_JOIN2 = pd.merge(df_votes_final,df_excel_le_plus_vote, on=['commune_code','N° de bureau de vote'])
```

```python
df_INNER_JOIN2
```

```python
df_INNER_JOIN2.to_excel('DataFrame pour les législatives (1er tour de 2012)/3)Les tendances de votes avec cordonnées pour les législatives (1er tour de 2012)_final_md.xlsx')
```

```python
df=df_INNER_JOIN.groupby(['commune_code', 'N° de bureau de vote'])['Nombre de voix du candidat'].idxmax()
# df_votes.groupby(['Code Commune (Insee Code)', 'N° de bureau de vote'])['Nombre de voix du candidat'].max()

list_index_le_plus_vote=df.to_frame()['Nombre de voix du candidat'].tolist()


df_le_plus_vote= pd.DataFrame(df_INNER_JOIN,
                  index=list_index_le_plus_vote)

df_le_plus_vote
```

```python
df_le_plus_vote.to_excel('Les tendances de votes avec cordonnées pour les législatives (1er tour de 2012).xlsx')
```
